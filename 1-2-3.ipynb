# Building a simulator in 2D

```python
# Perfect sensor
def sense(x):
    return x
```

### Euler integration

```python
def simulate(Δt, x, u):
    x += Δt * u
    return x
```

### Control policy

```python
def control(t, y):
    ### Circle Control Policy:
    ux = -sin(t)
    uy = cos(t)
    ### Ellipse Control Policy:
    #ux = -2*sqrt(3)*sin(t) - cos(t)
    #uy = -2*sin(t) + sqrt(3)*cos(t)
    ### Figure-Eight Control Policy:
    #ux = (sin(t)**2 - (cos(t)**2) + cos(t))/sqrt(2)
    #uy = (-(sin(t)**2) + cos(t)**2 + cos(t))/sqrt(2)
    ### Helix Control Policy:
    uz = 1
    return array([ux, uy, uz])
   
```

### Running the simulator

```python
tf = 10.
Δt = 0.1    # Time step
time = linspace(0.,tf, int(tf / Δt) + 1)  # Time interval


# Initial conditions
x = array([0., 0., 0.])
x_log = [copy(x)]

for t in time:
    y = sense(x)
    u = control(t, y)    
    x = simulate(Δt, x, u)
    x_log.append(copy(x))
    
x_log = array(x_log)
```

```python
grid()
fig = figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_log[:,0], x_log[:,1], x_log[:,2])
#plot(x_log[:,0], x_log[:,1])
```
